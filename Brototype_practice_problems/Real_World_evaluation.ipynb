##### Height Evaluation

```python
import matplotlib.pyplot as plt
```

```python
years = [2020, 2021, 2022, 2023, 2024]
height_cm = [ 120, 128, 135, 140, 142]

fig, ax = plt.subplots()
ax.plot(years, height_cm, linestyle=":", linewidth=2, color='red', marker='o')

ax.set_xlabel("Year")
ax.set_ylabel("Height (cm)")
ax.set_title("Height Over Years")
plt.show()
```

```python
subjects = ['Math', 'Science', 'English', 'Hindi', 'Malayalam']
marks = [85, 90, 78, 88, 92]

fig, ax= plt.subplots()
ax.bar(subjects, marks)
ax.set_xlabel("Subjects")
ax.set_ylabel("Marks")
ax.set_title("Marks by subject")

plt.show()
```

```python
hours_studied = [1, 2, 3, 4, 5, 6, 7]
marks_scored = [40, 45, 50, 60, 65, 80, 85]
fig, ax =plt.subplots()
ax.scatter(hours_studied, marks_scored)
ax.set_xlabel("Hours Studied")
ax.set_ylabel("Marks Scored")
ax.set_title("Study hours vs Marks")
plt.show()
```

```python
marks = [45, 50, 55, 60, 60, 62, 65, 70, 72, 75, 78, 80, 85, 88, 90, 95]
fig, ax = plt.subplots()
ax.hist(marks, bins=5)
ax.set_xlabel("Marks Range")
ax.set_ylabel("Number students")
ax.set_title("Distribution of Marks")
plt.show()
```

```python
ages = [5, 12, 18, 22, 25, 28, 30, 33, 35, 40, 42, 45, 50, 55, 60, 65, 70, 75, 80, 85]

fig, ax = plt.subplots()
ax.hist(ages, bins=5)
ax.set_xlabel("Age Range")
ax.set_ylabel("Number of people")
ax.set_title("Age Distribution in neighbourhood")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

normal_data = np.random.normal(50, 10, 1000)      # bell-shaped
skewed_data = np.random.exponential(10, 1000)      # skewed right

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].hist(normal_data, bins=20)
axes[0].set_title("Normal (Bell-shaped)")

axes[1].hist(skewed_data, bins=20)
axes[1].set_title("Skewed Right")

plt.show()
```
