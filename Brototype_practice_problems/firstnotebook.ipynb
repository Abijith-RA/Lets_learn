```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Test Plot: Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
```

```python
import sympy as sp

x = sp.Symbol('x')
expr = x**2 + 3*x + 2
derivative = sp.diff(expr, x)

print("Expression:", expr)
print("Derivative:", derivative)
```
