```python
days=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperature=np.array([30, 32, 31, 33, 29, 28, 30])
rainfall=np.array([0, 5, 2, 0, 10, 15, 3])
fig, axes = plt.subplots(2,1)
fig.suptitle("Weather Report")
axes[0]. plot(days, temperature)
axes[0].set_title("Temperature Graph")

axes[1]. bar(days, rainfall)
axes[1].set_title("Rainfall Graph")

plt.tight_layout()
plt.show()
```

```python
temp_this_week = np.array([30, 32, 31, 33, 29, 28, 30])
temp_last_week = np.array([28, 29, 30, 31, 30, 27, 29])
fig, ax= plt.subplots()
ax.plot(days, temp_this_week, label="This Week")
ax.plot(days, temp_last_week, label="Last Week ")

ax.set_xlabel("Day")
ax.set_ylabel("Temprature(°C)")
ax.set_title("Temperature Comparison")
ax.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temp_this_week = np.array([30, 32, 31, 33, 29, 28, 30])
temp_last_week = np.array([28, 29, 30, 31, 30, 27, 29])

fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(days, temp_this_week, label="This Week", linestyle="None", marker="s")
ax.plot(days, temp_last_week, label="Last Week", color="red", linestyle="--", marker="o")

ax.set_xlabel("Day")
ax.set_ylabel("Temperature (°C)")
ax.set_title("Temperature Comparison")
ax.legend()
plt.grid()
plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar']
your_marks = [70, 75, 80]
friend_marks = [60, 65, 90]
fig, ax = plt.subplots()
ax.plot(months, your_marks, label = "You", color = 'red', linestyle = '--', linewidth = 2, marker = "o")
ax.plot(months, friend_marks, label = "Friend", color = 'blue', linestyle = '-', linewidth = 2, marker = "s")
ax.set_xlabel("Month")
ax.set_ylabel("Marks")
ax.legend()
plt.grid()
plt.show()
```
