```python
import numpy as np

# We create an Unsigned array (uint8) with 4 numbers
unsigned_array = np.array([0, 127, 128, 255], dtype=np.uint8)

# We tell NumPy to look at the EXACT same memory using the Signed rule (int8)
signed_array = unsigned_array.view(np.int8)

print("Unsigned (uint8):", unsigned_array)
print("Signed (int8)  :", signed_array)
```

Array Operations

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
```

```python
#Arithmetic operations
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(b//a)
print(a>2)
```

```python
# Dot product
x = np.array([1, 2, 3])
y = np.array([1, 2, 3])
print(x*y)
print(np.dot(x,y))
```

```python
# Statistics
print(x.sum())
print("Mean",x.mean())
print("Min: ", x.min())
print("Max: ", x.max())
print("Std dev: ", x.std())
print("Variance: ",x.var())
m=np.array([[1,2,3],[4,5,6]])
print("Reshaped version: ", m.reshape(3,2))

print("Transpose:\n ", m.T)
```

```python
# function for element equality
print(np.array_equal(x,y))
```

```python
def cube_vectorized(arr):
    return arr ** 3
cube_vectorized(np.array([1,2,3,4]))
```
