#### Weather Story Telling



```python
import matplotlib.pyplot as plt
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
rainfall = [0, 5, 2, 0, 10, 15, 3]
colors = ['steelblue']*7
colors[5] = 'crimson'
```

```python
fig, ax = plt.subplots()
ax.bar(days, rainfall, color=colors)
ax.set_ylabel("Rainfall (mm)")
ax.set_title("Weekly Rainfall - Saturday Spike")
plt.show()
```
