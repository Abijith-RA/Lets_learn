```python
# 1. Enter the left-hand numbers: [[1, 2], [3, 4]]
A = np.array([[1, 2], [3, 4]])

# 2. Enter the right-hand totals: [5, 11]
b = np.array([5, 11])

# 3. Solve for x and y
solution = np.linalg.solve(A, b)
print(solution)
```

```python
import numpy as np
#5d MATRIX
# Method A: Create a 5D array of zeros
# Shape: 2x2x3x4x5 (Total elements = 2*2*3*4*5 = 240)
#array_zeros = np.zeros((2, 2, 3, 4, 5))

# Method B: Reshape sequential numbers into 5D
# Total elements must match the product of dimensions (2*2*2*2*2 = 32)
array_sequential = np.arange(54).reshape((3, 3, 3, 1, 2))
print(array_sequential)
```
