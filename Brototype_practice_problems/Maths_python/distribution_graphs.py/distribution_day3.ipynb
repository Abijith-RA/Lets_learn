```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
submission_times = [2, 3, 3, 4, 2, 3, 5, 3, 4, 40]
s=pd.Series(submission_times)

print("Mean", s.mean())
print("Median", s.median())
print("Mode", s.mode())
print("Skewness: ", s.skew())
```

```python
# Quantiles
Q1=s.quantile(0.25)
Q3=s.quantile(0.75)
IQR=Q3-Q1
lower_bound = Q1 - IQR*1.5
outer_bound = Q3 + IQR*1.5
outliers = s[(s<lower_bound)| (s>outer_bound)]
print("Q1: ", Q1, "Q3:", Q3, "IQR", IQR)
print("Bounds: ", lower_bound, outer_bound)
print("Outliers", outliers.tolist()) 
```
