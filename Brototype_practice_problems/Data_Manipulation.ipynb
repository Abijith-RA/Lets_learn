```python
import pandas as pd
```

```python
students = pd.DataFrame({
    'Name': ['Neha', 'Ravi', 'Sam', 'Anu'],
    'City': ['Kochi', 'Chennai', 'Kochi', 'Chennai'],
    'Score': [85, 90, 78, 65],
     'Age' : [20,18,21,23]
})

avg_by_city = students.groupby('City')['Score'].mean()
print(avg_by_city)
```

```python
print(students.groupby(['City', 'Name'])['Age'].median())
```

```python
df=pd.DataFrame({
    'Name': ['Neha','Sam', 'Ravi', 'Sam', 'Anu'],
    'City': ['Kochi', 'Kochi', 'Chennai', 'Kochi', 'Chennai'],
    'Score': [85, 90, 78,90, 65],
     'Age' : [20,21, 18,21,23],
     'Gender': ['F', 'M', 'M', 'M', 'F']})
print(df)
```

```python
print(df.isnull().sum())
```

```python
print("No of duplicate rows:", df.duplicated().sum())
```

```python
print(df.drop_duplicates())
```

```python
print(df.describe(include="all"))
```

```python
print(df.corr(numeric_only=True))
```

```python
print(df['Score'].mode())
```

```python
print(df.groupby('City')['Score'].agg(['max', 'mean']))
```

DATA TRANSFORMATIONS

```python
df['Percent of marks']= df['Score']*(90/100)
```

```python
print(df)
```

```python
df['Grade']=df['Score'].apply(lambda x: 'Pass' if x>=85 else 'Fail')
#print(df)
```

```python
df['Age']= df['Age'].astype(int)
```

```python
df.rename(columns={'Marks': 'Score', 'Years': 'Age'}, inplace=True)
```

```python
df['Gender']=df['Gender'].map({'M': 'Male', 'F': 'Female'})
```

```python
print(df['Gender'])
```

```python
pd.crosstab(df['City'], df['Score'])
```

```python
df.groupby('City')['Score'].mean()
```

```python
print("Std dev: ",df['Score'].std())
print("Variance: ", df['Score'].var())
```

```python
df.corr(numeric_only=True)
```

```python
city_stats= df.groupby('City')['Score'].agg(['mean', 'count'])
print(city_stats)
```

```python
print(city_stats.sort_values('mean', ascending=False))
```
